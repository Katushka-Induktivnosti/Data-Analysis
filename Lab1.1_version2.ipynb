```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

size = 50

a = -0.75
b = 4
x = np.random.random(size=size)
noise = np.random.random(size=size)

y = a*x+b+noise

f = plt.figure(figsize=(10,10))
c = f.add_subplot(111)
c.scatter(x, y, marker='.')
c.set_aspect('equal')

data = {'X': x, 'Y': y}
df = pd.DataFrame(data, columns = ['X','Y'])
X = df['X'].values
Y = df['Y'].values

mean_x = np.mean(X)
mean_y = np.mean(Y)

data = df

m = len(X)
numer = 0
denom = 0
for i in range(m):
  numer += (X[i] - mean_x) * (Y[i] - mean_y)
  denom += (X[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)

print (f'a_predicted = {m} \na = {a} \nb_predicted = {c} \nb = {b}')

max_x = np.max(X)+1
min_x = np.min(Y)-4

x = np.linspace (min_x, max_x, 10)
y = c + m * x

plt.plot(x, y, c='green', label='Regression Line')
plt.scatter(X, Y, c='orange', label='data points')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
